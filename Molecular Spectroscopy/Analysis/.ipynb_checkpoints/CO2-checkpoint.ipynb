```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

size = [30,10]
path = '/mnt/c/Users/alon4/Dropbox/Labs/Lab C1/Molecular Spectroscopy'
outside = pd.read_excel(os.path.join(path,'Measurements/CO2/gas co2 outside.xlsx'))
# outside.set_index(outside.columns[0],inplace=True) # for pandas use
outside = outside.to_numpy().T # for numpy use
outside[1] -= min(outside[1])

breath = pd.read_excel(os.path.join(path,'Measurements/CO2/gas co2 blow.xlsx'))
# breath.set_index(breath.columns[0],inplace=True) # for pandas use
breath = breath.to_numpy().T # for numpy use
breath[1] -= min(breath[1]) # so that min(Y) = 0

def TrapIntegral(X, Y):
    return sum([(Y[i]+Y[i+1])*np.abs((X[i+1]-X[i]))/2 for i in range(len(Y)-1)])
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [30,10]
img = mpimg.imread(os.path.join(path,'Literature/CO2 Transmitance.png'))
imgplot = plt.imshow(img)
plt.show()
plt.plot(outside[0], outside[1])
plt.show()
plt.plot(breath[0], breath[1])
plt.show()
```

```python
CO2_O = outside[:,(outside[0] < 2400) & (outside[0] > 2250)]
CO2_B = breath[:,(breath[0] < 2400) & (breath[0] > 2250)]
plt.plot(CO2_O[0], CO2_O[1])
plt.show()
plt.plot(CO2_B[0], CO2_B[1])
plt.show()
```

```python
area_O = TrapIntegral(CO2_O[0],CO2_O[1])
area_B = TrapIntegral(CO2_B[0],CO2_B[1])
print(area_B/area_O)
```
