```python
import pandas as pd
from bisect import bisect
import numpy as np
import eddington
from matplotlib import pyplot as plt
import matplotlib as mpl
from scipy.interpolate import CubicSpline
from scipy.optimize import curve_fit
plt.rcParams['figure.figsize'] = [10,7.5]
plt.rcParams.update({'font.size': 18, 'font.weight':'bold'})
plt.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'
```

```python
low = pd.read_excel('B=5_T=1.7_Vg=-0.5to0.5.xlsx').set_index('Time (Second)')
low = np.array([low['Gate Voltage (Volt)'].to_list(),low['SR510 No. 1 (Ohm}'].to_list(),low['SR510 No. 2 (Ohm}'].to_list()])
high = pd.read_excel('B=5_T=4.2_Vg=0.5to-0.5.xlsx').set_index('Time (Second)')
high = np.array([high['Gate Voltage (Volt)'].to_list(),high['SR510 No. 1 (Ohm}'].to_list(),high['SR510 No. 2 (Ohm}'].to_list()])
```

```python
plt.plot(low[0],low[1], label='1.7 Kelvin')
plt.plot(high[0],high[1], label='4.2 Kelvin')
plt.ylabel(r'$\text{R}_{\boldsymbol{xx}}$ [$\boldsymbol{\Omega}$]')
plt.xlabel(r'$\text{V}_{\boldsymbol{g}}$ [Volt]')
plt.title(r'$\text{R}_{\boldsymbol{xx}}$ as a function of Gate Voltage' + 
          '\nat a Magnetic Field of 5 Tesla for both Temperatures')
plt.legend()
plt.savefig('RxxAt5Tesla.png')
plt.show()

plt.plot(low[0],low[2], label='1.7 Kelvin')
plt.plot(high[0],high[2], label='4.2 Kelvin')
plt.ylabel(r'$\text{R}_{\boldsymbol{xy}}$ [$\boldsymbol{\Omega}$]')
plt.xlabel(r'$\text{V}_{\boldsymbol{g}}$ [Volt]')
plt.title(r'$\text{R}_{\boldsymbol{xy}}$ as a function of Gate Voltage' + 
          '\nat a Magnetic Field of 5 Tesla for both Temperatures')
plt.legend()
plt.savefig('RxyAt5Tesla.png')
plt.show()
```
