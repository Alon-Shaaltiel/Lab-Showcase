```python
import pandas as pd
import eddington
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams['figure.figsize'] = [16,9]
plt.rcParams.update({'font.size': 20, 'font.weight':'bold'})
plt.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'
```

```python
data = eddington.FittingData.read_from_excel(filepath='B=0.5_T=4.2_Vg=0.5to-0.5.xlsx', sheet='FitData')
result = eddington.fit(data, eddington.linear)
result.save_txt("Lin Fit/lin_result.txt")
eddington.plot_fitting(eddington.linear, data, dict(a = result.a),title_name='')
plt.title('Inverse to Resistance perpendicular to current stream\n as a function of Gate Voltage on capacitor')
plt.xlabel(r'Gate Voltage [Volt]')
plt.ylabel(r'$\text{R}_{\boldsymbol{\text{xy}}}^{\boldsymbol{-1}}$ [$\boldsymbol{\Omega^{-1}}$]')
plt.legend(('Fit','Data'))
plt.savefig("Lin Fit/Fit.png")
plt.show()
```

```python
df = pd.read_excel("Fit Data.xlsx", sheet_name='Python')
def Metal(a,x):
    return a[0] + a[1]*x**3 + a[2]*x**5

def SemiConductor(a,x):
    return a[0]+a[1]*np.exp(a[2]/x)
    
aM = [46.9, 1.001*10**(-4),3.134e-9]
aSC = [72.010, 1.26e-6, 2*1707]
rM = Metal(aM, df['Temperature (K)'])
rSC = SemiConductor(aSC, df['Temperature (K)'])

plt.plot(df['Temperature (K)'], 1/df['Resistance (Ohm)'])
plt.plot(df['Temperature (K)'], (1/rM + 1/rSC))
plt.title('Inverse of Total Resistance as a function of Sample Temperature')
plt.xlabel(r'Temperature [Kelvin]')
plt.ylabel(r'$\text{R}_\text{xx}^{\boldsymbol{-1}}$ [$\boldsymbol{\Omega^{-1}}$]')
plt.legend(('Measurements','Fit'))
plt.savefig("TotalResistance_vs_Temp.png")
plt.show()
```
