An example using an out-of-the-box function and an implemented function 

```python
import numpy as np
import eddington
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
ex_func1 = lambda x: np.sin(x + 0.5*np.random.rand()) + 0.5*np.random.rand()
x = np.linspace(np.pi,6*np.pi,num=200)
dx = np.zeros_like(x) + 0.3
y1 = np.array([ex_func1(i) for i in x])
dy1 = 0.05*y1

plt.rcParams['figure.figsize'] = [30,10]
plt.errorbar(x,y1,xerr=dx,yerr=dy1,fmt='.')
plt.show()
```

```python
data1 = eddington.FittingData(dict(a = x, da = dx, V = y1, dV = dy1))
# Eddington expects a dictionary with 4 (?) keys. 
# Each key contains an iterable. The keys should be ordered by x, dx, y, dy.

# We'll fit a sine function to the data so we have to make an initial guess.
# The sine function's syntax is: a[0] * sin(a[1] * x + a[2]) + a[3]
# We can view sine's syntax using eddington.sin.syntax

guess1 = [(max(y1)-min(y1))/2, 1, 0.5, 0]
result1 = eddington.fit(data1, eddington.sin, a0=guess1)
eddington.plot_fitting(eddington.sin, data1, dict(a = result1.a), title_name='Plot')
# eddington.plot_residuals(eddington.sin, data1, dict(a = result1.a), title_name='Residuals') # DOESNT WORK???
plt.xlabel('x [cm]') # Here you may want to change the font and size or use latex
plt.ylabel('y [V]')
plt.legend(('fit','data'))
plt.savefig("SineFit.png")
plt.show()

result1.save_txt("SineResults.txt")
print(result1)
```

```python
# We can now interact with the fit parameters and statistical metrics
print(f'''
{result1.a}
{result1.aerr}
P-probability is{'' if (result1.p_probability < 0.95) and (result1.p_probability > 0.05) else ' not'} in the desired range''')
```

```python
noisyFunc = lambda x: 3*np.sin(x+0.1*np.random.rand())/(x**2 + 8 + np.random.rand())
y2 = np.array([noisyFunc(i) for i in x])
dy2 = 0.05*y2

plt.errorbar(x,y2,xerr=dx,yerr=dy2,fmt='.')
plt.show()
```

```python
# Exporting to excel, we will later import into Eddington
import pandas as pd
df = pd.DataFrame(np.vstack((x,dx,y2,dy2)).T, columns=['x [cm]','Δx [cm]', 'y [V]', 'Δy [V]'])
df.set_index(df.columns[0],inplace=True)
df.to_excel('SampleTest.xlsx')
```

```python
@eddington.fitting_function(n=5, syntax="(a[0] * np.sin(a[1]*x+a[2])) / (x**2+a[3]) + a[4]")
def funcToFit(a, x):
    return (a[0] * np.sin(a[1]*x+a[2])) / (x**2+a[3]) + a[4]
```

```python
data2 = eddington.FittingData.read_from_excel(filepath='SampleTest.xlsx', sheet='Sheet1')
guess2 = [2.65,0.96,0.001,7.65,0.001]
result2 = eddington.fit(data2, funcToFit, a0=guess2)
eddington.plot_fitting(funcToFit, data2, dict(a = result2.a), title_name='Plot')
# eddington.plot_residuals(...) for residuals
# DOESNT WORK???
plt.xlabel('x [cm]')
plt.ylabel('y [V]')
plt.legend(('fit','data'))
plt.savefig("ExampleFit.png")
plt.show()

result2.save_txt("ExampleResults.txt")
print(result2)
```
