```python
from scipy.io import loadmat
from tqdm.auto import tqdm
data = loadmat('Electrons_loose_medium_tight.mat')
```

```python
# DONT USE THIS, IT SUCKS!
index = 0
exxc = np.zeros((data['Electrons'][0][0][0][0][0],len(data['Electrons'][0][0])))
for i in range(len(exxc[:])):
    for j in range(1,len(exxc[0])):
        if j==1:
            exxc[i,j-1] = index
        
        exxc[i][j] = data['Electrons'][0][0][j][0][i]

elcM = 0.511
MassSquared = [[],[],[]] # Loose Medium Tight
for k in tqdm(range(1,len(data['Electrons'][0]))):
    index += 1
    temp = np.zeros((data['Electrons'][0][k][0][0][0],len(data['Electrons'][0][0])))
    for i in range(len(temp[:])):
        for j in range(1,len(temp[0])):
            if j==1:
                temp[i,j-1] = index

            temp[i][j] = data['Electrons'][0][k][j][0][i]

    exxc = np.vstack((exxc,temp)) # index eta phi Pt charge loose medium tight 
    for a in range(len(temp[:])):
        for b in range(a+1,len(temp[:])):
            if temp[a][4]+temp[b][4]==0:
                momentsA = np.array([temp[a][3]*np.cos(temp[a][2]), temp[a][3]*np.sin(temp[a][2]), temp[a][3]*np.sinh(temp[a][1])])
                momentsB = np.array([temp[b][3]*np.cos(temp[b][2]), temp[b][3]*np.sin(temp[b][2]), temp[b][3]*np.sinh(temp[b][1])])
                MinvarSQRD = 2*elcM**2+2*np.sqrt(elcM**2+momentsA.dot(momentsA))*np.sqrt(elcM**2+momentsB.dot(momentsB))-2 * momentsA.dot(momentsB)
                if temp[a][-3] and temp[b][-3]:
                    MassSquared[0].append(MinvarSQRD)
                if temp[a][-2] and temp[b][-2]:
                    MassSquared[1].append(MinvarSQRD)
                if temp[a][-1] and temp[b][-1]:
                    MassSquared[2].append(MinvarSQRD)
            
MassSquared[0] = np.sqrt(np.array(MassSquared[0])) * 10**(-3)# Loose
MassSquared[1] = np.sqrt(np.array(MassSquared[1])) * 10**(-3)# Medium
MassSquared[2] = np.sqrt(np.array(MassSquared[2])) * 10**(-3)# Tight
Mass_Copy = MassSquared.copy()
MassSquared[0] = MassSquared[0][(MassSquared[0] < 150) & (MassSquared[0] > 1)]
MassSquared[1] = MassSquared[1][(MassSquared[1] < 150) & (MassSquared[1] > 1)]
MassSquared[2] = MassSquared[2][(MassSquared[2] < 150) & (MassSquared[2] > 1)]
```

```python
eventsUps = np.array(eventsUps)
mask = eventsUps>max(eventsUps)/2 # [False False False True ... True True False False ...]
i = [0,len(mask)]
while not mask[i[0]]:
    i[0]+=1
while not mask[::-1][i[1]]:
    i[1]-=1

DX = [(binsUps[i+1]-binsUps[i])/2 for i in range(len(binsUps)-1)]
X = [(binsUps[i+1]+binsUps[i])/2 for i in range(len(binsUps)-1)]
DY = [np.sqrt(eve) for eve in eventsUps]
Data = np.array([X,DX,eventsUps,DY]).T
df = pd.DataFrame(Data[Data[:,2] > 0], columns=['Mass [GeV]', 'Mass [Gev]', 'events', 'events']).set_index('Mass [GeV]')
# df.to_excel(excel_writer = os.path.join(path, "Fit/Third Part/CR_UPS/CR_UPS.xlsx"))
df.to_excel(excel_writer = os.path.join(path, "Fit/Third Part/SR_UPS/SR_UPS.xlsx"))
```
